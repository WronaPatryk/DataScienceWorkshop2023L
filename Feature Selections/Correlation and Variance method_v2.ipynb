#### 1. Correlation Method

```python
plt.figure(figsize = (16, 9))
sns.heatmap(df.corr())
```

```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr

# Check those cols who has more than 80% correlation
corr_features= correlation(df, 0.9)
corr_features
```

# VIF feature selection

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def leaveKusingVIF_QUICK(data, k):
    while k < len(data.columns):
        ds=pd.Series([variance_inflation_factor(data.values, i)   
           for i in range(data.shape[1])],   
              index=data.columns)  
        col_to_be_deleted = data.columns[np.argmax(ds)]

        data = data.drop(col_to_be_deleted,axis=1)
    return data


def calculateVIFpercentages(vif_left_cols, keyboard, mouse, both):
    
    b_vif= np.array([b in vif_left_cols for b in both])
    k_vif = np.array([b in vif_left_cols for b in keyboard])
    m_vif = np.array([b in vif_left_cols for b in mouse])

    # percentages:

    b_vif_p = b_vif.sum()/len(b_vif) * 100
    k_vif_p = k_vif.sum()/len(k_vif) * 100
    m_vif_p = m_vif.sum()/len(m_vif) * 100
    
    return b_vif_p, k_vif_p, m_vif_p


```

```python
# import dataset
df = pd.read_csv("../data/data.csv")
df = df.iloc[:, 4:]
df.head(2)

# dropping 0 variance columns
df = df.drop(df.columns[np.unique(np.where(df.var() == 0))], axis = 1)

### Check the each missing value of col in percentage
null_var = df.isnull().sum()/df.shape[0] * 100
null_var
# drop the col which have more than 50% missing values
drop_columns = null_var[null_var > 50].keys()
drop_columns 



df = df.drop(columns = drop_columns)
```

```python
import miceforest as mf
```

```python
kds = mf.ImputationKernel(df, random_state=44)

# run for 2 iterations
kds.mice(2, n_estimators=30)
# Return the completed dataset.
df = kds.complete_data()
```

```python

b2 = []
k2 = []
m2 = []

n_cols = len(df.columns)

for t in range(0, n_cols - 10, 5):
    
    ts = time.time()
    print(f'Iteration: {t}')
    filtered_data = leaveKusingVIF_QUICK(df, n_cols - t)
    
    left_cols = list(filtered_data.columns)

    b,k,m = calculateVIFpercentages(left_cols, keyboard, mouse, both)
    
    b2.append(b)
    k2.append(k)
    m2.append(m)
    
    print(f'--- time {round(time.time() - ts, 2)} s')





```
