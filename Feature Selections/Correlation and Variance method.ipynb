### Feature Selection

#### 1. Correlation Method

```python
plt.figure(figsize = (16, 9))
sns.heatmap(df.corr())
```

```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr

# Check those cols who has more than 80% correlation
corr_features= correlation(df, 0.9)
corr_features
```

```python
df = df.drop(corr_features,axis=1)
df.shape
```

### 2. Removing features with low variance

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
df = pd.read_csv("data.csv")
df = df.iloc[:, 4:]
df.head(2)
df.shape
```

```python
sel = VarianceThreshold(threshold=(.8 * (1 - .8)))
sel.fit_transform(df)
```

```python
sel.get_support()
```

```python
# True: High Variance
# False: Low Variance
```

```python
concol = [column for column in df.columns 
          if column not in df.columns[sel.get_support()]]
```

#### Picking Up the low Variance Columns:

```python
for features in concol:
    print(features)
```

```python
df.drop(concol,axis=1)
```
