```python
import numpy as np, pandas as pd
```

```python

PATH_ = "feature_sets/"

def getFeaturesAsLists():
    
    list_of_lists = []
    
    
    # @@@@@@@@@@@@@@@@@@@ IMPUTATION
    imputation_path = PATH_ + "imp_selected.csv"
    imp_list = pd.read_csv(imputation_path, header = None).values
    resulting_features = []
    for ll in imp_list:
        for el in ll[1:]:
            resulting_features.append(el)
    uniq = np.unique(resulting_features, return_counts = True)
    # counts
    uniq[1][np.argsort(uniq[1])]
    # the best 23 features:
    best_23_features_VIF = uniq[0][np.argsort(uniq[1])][0:23]
    best_23_features_VIF
    list_of_lists.append(best_23_features_VIF)
    # @@@@@@@@@@@@@@@@@@@
    return list_of_lists
    
    
def aggregateLists(list_of_lists):
    
    final_list = []
    
    for list_ in list_of_lists:
        
        for el in list_:
            final_list.append(el)
    uniq = np.unique(final_list, return_counts = True)
    # counts
    uniq[1][np.argsort(uniq[1])]
    # the best 23 features:
    best_features = uniq[0][np.argsort(uniq[1])][0:23]
    return best_features
    
    
```

```python
list_of_lists = getFeaturesAsLists()
res = aggregateLists(list_of_lists)

```
