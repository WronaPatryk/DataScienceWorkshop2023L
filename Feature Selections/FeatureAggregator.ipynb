```python
import numpy as np, pandas as pd
```

```python

PATH_ = "feature_sets/"

def getFeaturesAsLists():
    
    list_of_lists = []
    
    
    # @@@@@@@@@@@@@@@@@@@ IMPUTATION
    imputation_path = PATH_ + "imp_selected.csv"
    imp_list = pd.read_csv(imputation_path, header = None).values
    resulting_features = []
    for ll in imp_list:
        for el in ll[1:]:
            resulting_features.append(el)
    uniq = np.unique(resulting_features, return_counts = True)
    # counts
    uniq[1][np.argsort(uniq[1])]
    # the best 23 features:
    best_23_features_VIF = uniq[0][np.argsort(uniq[1])][0:23]
    best_23_features_VIF
    list_of_lists.append(best_23_features_VIF)
    # @@@@@@@@@@@@@@@@@@@
    
    # @@@@@@@@@@@@@@@@@@@ PAIR-WISE CORRELATION
    _path = PATH_ + "Corr_sel_features.csv"
    corr_df = pd.read_csv(_path)
    resulting_features = np.array(corr_df.columns)
    list_of_lists.append(resulting_features)
    
    # @@@@@@@@@@@@@@@@@@@ PAIR-WISE PCA
    _path = PATH_ + "PCA_Sel_Feature.csv"
    pca_df = pd.read_csv(_path)
    resulting_features = np.array(pca_df.columns)
    list_of_lists.append(resulting_features)
    
    # @@@@@@@@@@@@@@@@@@@ PAIR-WISE VARIANCE
    _path = PATH_ + "Var_Sel_features.csv"
    var_df = pd.read_csv(_path)
    resulting_features = np.array(var_df.columns)
    list_of_lists.append(resulting_features)
    
    
    
    
    return list_of_lists
    
    
def aggregateLists(list_of_lists, n_of_features=20):
    
    final_list = []
    
    for list_ in list_of_lists:
        
        for el in list_:
            final_list.append(el)
    uniq = np.unique(final_list, return_counts = True)
    
#     return uniq
    # counts
#     uniq[1][np.argsort(uniq[1])]
    # the best 23 features:
    print(uniq[0])
    print(uniq[1][np.argsort(uniq[1])])
    
    
    best_features = np.flip(uniq[0][np.argsort( uniq[1] )])[0:n_of_features]
    
    counts = np.flip(uniq[1][np.argsort( uniq[1] )])[0:n_of_features]
    print(counts)
    
    
    aux = pd.DataFrame(best_features)
    aux["counts"] = counts
    aux.rename(columns = {0:'feature', 'counts' : 'counts'}, inplace = True)
    
    return aux
    

```

```python
list_of_lists = getFeaturesAsLists()
res = aggregateLists(list_of_lists, 29) # or 29

res
```
