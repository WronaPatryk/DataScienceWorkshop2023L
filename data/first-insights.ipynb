```python
data = pd.read_csv("data.csv")
```

```python
import datetime
ts = data.event_time[0]
def to_unix_time(ts):
    epoch = datetime.datetime.utcfromtimestamp(0) # start of epoch time
    my_time = datetime.datetime.strptime(ts, "%Y-%m-%d %H:%M:%S.%f")
    delta = my_time - epoch
    return(delta.total_seconds() * 1000.0)

to_unix_time(ts)
```
